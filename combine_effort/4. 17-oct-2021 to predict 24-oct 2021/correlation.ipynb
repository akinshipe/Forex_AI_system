```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
#import matplotlib.pyplot as plt
from sklearn import metrics
import numpy as np
```

```python
currency_pairs = ['USDCHF10080',
                 'GBPUSD10080', 'EURUSD10080', 'USDJPY10080', 'USDCAD10080', 'AUDUSD10080', 'NZDUSD10080',
                 'GBPCHF10080', 'EURCHF10080', 'CHFJPY10080', 'CADCHF10080', 'AUDCHF10080', 'NZDCHF10080', 'EURGBP10080',
                 'GBPJPY10080', 'GBPCAD10080', 'GBPAUD10080', 'EURJPY10080', 'EURCAD10080', 'EURAUD10080', 'EURNZD10080',
                'CADJPY10080', 'AUDJPY10080', 'NZDJPY10080', 'AUDCAD10080', 'NZDCAD10080', 'AUDNZD10080'
                ]

headers = ['date', 'ignore', 'open', 'high', 'low', 'close', 'volume'  ]
```

```python
writer = pd.ExcelWriter('files/auto-correlation.xlsx')



for z in currency_pairs:
    
    current_currency = z.replace('10080','')
    raw_data = pd.read_csv('files/' + current_currency + '.csv'  ) 

    currency_list = []
    ten_weeks = []
    fifteen_weeks = []
    twenty_weeks = []
    thirty_weeks = []
    fifty_weeks = []


    for x in currency_pairs: 
        
        currency_list.append(x.replace('10080',''))
        
        weeks = [10, 15, 20, 30, 50 ]

        for y in weeks:
            
            current_data = raw_data.copy()
           

            current_data = current_data.head(y)

            new_df = current_data[['nextweek_'+ current_currency, x.replace('10080','')]]

            
            correlation = new_df.corr()
            correlation = correlation[x.replace('10080','')][0:1].values[0]
            correlation = abs(round(correlation, 5))
            #print(correlation)
            
            if y == 10:
                ten_weeks.append(correlation)
            
            if y == 15:
                fifteen_weeks.append(correlation)
                
            if y == 20:
                twenty_weeks.append(correlation)
                
            if y == 30:
                thirty_weeks.append(correlation)
                
            if y == 50:
                fifty_weeks.append(correlation)
                

    current_currency_df = pd.DataFrame()
    
    
    current_currency_df[1]= currency_list
    current_currency_df['fifty']= fifty_weeks
    
    current_currency_df[2]= ''
    current_currency_df[3]= currency_list
    current_currency_df['thirty']= thirty_weeks
    
    current_currency_df[4]= ''
    current_currency_df[5]= currency_list
    current_currency_df['twenty']= twenty_weeks
    
    current_currency_df[6]= ''
    current_currency_df[7]= currency_list
    current_currency_df['fifteen']= fifteen_weeks
    
    
    current_currency_df[8]= ''
    current_currency_df['currency']= currency_list
    current_currency_df['ten']= ten_weeks
    
    
    print(current_currency, '---------------------------------------------------------------')
    #print(current_currency_df)
    
    current_currency_df.loc[current_currency_df.shape[0]] = ''
    current_currency_df.loc[current_currency_df.shape[0]] = ''
    current_currency_df.loc[current_currency_df.shape[0]] = ''
    
       
    current_currency_df = current_currency_df.shift()
    current_currency_df = current_currency_df.shift()
    current_currency_df = current_currency_df.shift()
    
    cor = ['', 0.279, '', '', 0.361, '', '', 0.444, '', '', 0.514, '', '',0.632]
    
    current_currency_df.iloc[1] = cor

    current_currency_df.to_excel(writer, current_currency, index = False )
    
                

writer.close()
```
