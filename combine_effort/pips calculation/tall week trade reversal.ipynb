```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
#import matplotlib.pyplot as plt
from sklearn import metrics
import numpy as np
import time
from datetime import datetime
```

```python
for root, dirs, files in os.walk("../", topdown=False):
    pass

print(len(dirs))
```

```python
all_week_dataframes = []



for a in dirs:
    
    path = '../'+ a + '/files/maximaldraw_down_and_up.csv'
    
    if os.path.exists(path) == False:
        
        print(path, ' not_exist')
        continue
        
        
    current_week_dataframe = pd.read_csv(path )
    
    
    
    # reverses dataframe so that iterrows walk through the rows from monday to friday
    current_week_dataframe =  current_week_dataframe.iloc[::-1]
    
    
    
    current_week_dataframe = current_week_dataframe.reset_index(drop=True)
    
    
    
    current_week_dataframe = current_week_dataframe.iloc[: , :2]
    
    #print(current_week_dataframe)
    
    
    
    
    current_dt_lenght = len(current_week_dataframe.index)
    
    if current_dt_lenght > 240:
        print(path,)
        raise Exception('Rows are not supposed to me more than 240')
        
        
     #print(current_week_dataframe)
    # this is to removes weeks that have holiday or missing data in them
    if current_dt_lenght <230:
        
        print(path, ' row is less than 230')
        continue
        
    
    current_week_data = {'week_path' : path, 'dataframe' : current_week_dataframe,  }
     
    all_week_dataframes.append(current_week_data)
    
    

    
        
print(len(all_week_dataframes))
        
        
```

```python
for n in all_week_dataframes:
   
    current_week_dataframe = n['dataframe']
    
    def invert_profit(value):
        return value*-1
    
    current_week_dataframe['profit'] = current_week_dataframe['profit'].apply(invert_profit)
    
    n['dataframe'] =    current_week_dataframe     
    

print( all_week_dataframes)
```

```python

statistic_df = pd.DataFrame()

start_date_index_list =[]

cut_off_date_index_list =[]

draw_down_list = []

total_spread_list = []

maximal_draw_down_list = []

profit_list=[]

profit_ratio_list=[]





start_date_index = [f+1 for f in range(0, 230)]

draw_down = [f*-1 for f in range(100, 3000, 100)] 

total_spread = -200

cut_off_date_index =   [f+1 for f in range(0, 240)]


for t in start_date_index:

    for g in cut_off_date_index:
        
        if g + t > 242:
            
            print( 'last row reached by cutoff' )
            continue

        for q in draw_down :

            maximal_draw_down = total_spread + q

            week= []

            profit =[]
            
            for b in all_week_dataframes:

                week_name = 'week_' + b['week_path'][3:7]

                current_week_dataframe = b['dataframe'].copy(deep=True)

                current_week_dataframe =  current_week_dataframe.iloc[t:]

                current_week_dataframe = current_week_dataframe.reset_index(drop=True)



                #print(start_value)



                current_week_dataframe = current_week_dataframe.head(g +1 )
                
                
                
                

                start_value = current_week_dataframe.head(1)['profit'][0]

                #print( current_week_dataframe)

                current_profit = current_week_dataframe['profit'].tail(1).values[0] - start_value

                #print(current_profit)

                for index, row in current_week_dataframe.iterrows():

                    if row.profit - start_value <= maximal_draw_down:

                        current_profit = maximal_draw_down

                        #print(week_name)

                        break

                
                
                
                profit.append(current_profit) 

                week.append(week_name)
                
                
            start_date_index_list.append(t)
                
            cut_off_date_index_list.append(g)

            draw_down_list.append(q)

            total_spread_list.append(total_spread)
            
            maximal_draw_down_list.append(maximal_draw_down)
            
            profit_list.append(sum(profit))
            
            profit_ratio_list.append(sum(profit)/(10*maximal_draw_down*-1))

            

            #print(profit)        
            #print(sum(profit) )

#            print('cut off date :  ',g ,sum(profit),  len(profit), 'draw_down :', q , ' max_draw_down  : ', 
#                  maximal_draw_down, ' risk/reward  ', sum(profit)/(10*maximal_draw_down*-1), 'weeks: ',len(week )) 
#        print('--------------------------------------------------------------------------------------------------------')

        print('cut off index ', g)
    print('start date index : -------------------------------------------------------------------------------', t)
```

```python
statistic_df['start_date_index'] = start_date_index_list


statistic_df['cut_off_date_index'] = cut_off_date_index_list

statistic_df['total_spread'] = total_spread_list

statistic_df['draw_down'] = draw_down_list

statistic_df['maximal_draw_down'] = maximal_draw_down_list



statistic_df['profit'] = profit_list

statistic_df['profit_ratio'] = profit_ratio_list





print(statistic_df)
```
