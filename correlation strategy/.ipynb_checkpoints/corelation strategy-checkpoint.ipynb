```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
#import matplotlib.pyplot as plt
from sklearn import metrics
import numpy as np
import time
from datetime import datetime, timedelta

all_currency_pairs = ['USDCHF',
                 'GBPUSD', 'EURUSD', 'USDJPY', 'USDCAD', 'AUDUSD', 'NZDUSD',
                 'GBPCHF', 'EURCHF', 'CHFJPY', 'CADCHF', 'AUDCHF', 'NZDCHF', 'EURGBP',
                 'GBPJPY', 'GBPCAD', 'GBPAUD', 'EURJPY', 'EURCAD', 'EURAUD', 'EURNZD',
                'CADJPY', 'AUDJPY', 'NZDJPY', 'AUDCAD', 'NZDCAD', 'AUDNZD'
                ]



headers = ['date', 'ignore', 'open', 'high', 'low', 'close', 'volume'  ]
```

```python
raw_df = pd.read_csv("../raw_weekly_data/NZDCAD10080.csv", names = headers )



needed_weeks_df = raw_df.loc[(raw_df['date'] >= "2015.01.01")   ]   

needed_weeks_df = needed_weeks_df.iloc[::-1]

needed_weeks_df = needed_weeks_df.reset_index(drop=True)



available_weeks = needed_weeks_df['date'].values



#print(raw_df)
print(available_weeks)
```

```python
for y in available_weeks:
    
    current_date = datetime.strptime(y, '%Y.%m.%d').strftime('%A')
    
    if current_date != 'Sunday':
       
        print('error-----------------------------', y, '  is not a sunday')
        
        raise Exception('error-----------------------------', y, ' did is not a sunday')
        
    
```

```python
for a in available_weeks:
  
  
    path = "./files/" + a

    try: 
        os.makedirs(path)
        control_file = pd.DataFrame()
        
        for  b in range(8):
            control_file['control' + str(b)] = -1
            
            
        control_file.to_csv(path + '/control.csv')
        
        

    except OSError as error: 
         raise Exception(error)
 
```

```python
# classif historical currency data

raw_currency_pair_classification_df = []

for a in all_currency_pairs:
    
    classification_df = pd.DataFrame()
    
    classification_score =[]
    
    current_currency_df = pd.read_csv("../raw_weekly_data/" + a + "10080.csv", names = headers ) 
        
        
    current_currency_df = current_currency_df.loc[(current_currency_df['date'] >= '2014.01.01' ) ]
    
    current_currency_df = current_currency_df.iloc[::-1]   
        
    current_currency_df = current_currency_df.reset_index(drop=True)
    
    classification_df['date'] = current_currency_df['date']
    
    
    for b in (range(len(current_currency_df.index))):
            
            row = current_currency_df.loc[b,:]
                            
            #condition when the curreny movement is a buy
            

            if row.open < row.close:
                
                 classification_score.append(1)
                #print(b, a, row.date, row.open, row.close)
                          
            else:
                
                #print(b, a,row.date, row.open, row.close)
                 classification_score.append(-1)

                    
    
    classification_df['class'] = classification_score
    
    raw_currency_pair_classification_df.append({'name': a, 'data': classification_df })
                    
                    
                    
    #print('------------------------------------------------------------------------------------------',a)
    
    
print(raw_currency_pair_classification_df)
```

```python


for c in available_weeks :
    
    current_week_classification_df = pd.DataFrame()

    for index, d in enumerate(raw_currency_pair_classification_df):
              
        current_currency = d['name']

        current_currency_df = d['data']
        #print(d)
        
        
        current_currency_df = current_currency_df.loc[(current_currency_df['date'] <= c ) ]
        
        current_currency_df = current_currency_df.reset_index(drop=True)
                
        current_currency_df = current_currency_df.head(52)   
         
        #print(c, d['name'],current_currency_df)  
        
        
        if index == 0 :
            
            current_week_classification_df['date'] = current_currency_df['date']
        
    
        current_week_classification_df[current_currency] = current_currency_df['class']
    
    
    
    #print(current_week_classification_df)   

    print('-----------------------------------------------------------------------------------------------------' , c )

    
    # creating dataframe with next week for each currency
    
    for index, e in enumerate(all_currency_pairs):
        
        
        current_week_classification_dataframe = current_week_classification_df.copy()
        
        current_currency_list = current_week_classification_dataframe[e].values.tolist()
        
        current_currency_list.insert(0,0)
        
        current_currency_list.pop()
        
        
        
        current_week_classification_dataframe['nextweek_'+ e] = current_currency_list
        
        if index == 0:
        
            currrent_week_values_df = current_week_classification_dataframe.iloc[:1, :28]


            current_week_values_df = currrent_week_values_df.reset_index(drop =True)

            current_week_values_df.to_csv("files/" + c + "/current_week_values.csv", index = False)

            #print(current_week_values_df)
        
        
        print(current_week_classification_dataframe)
        
        
        current_week_classification_dataframe = current_week_classification_dataframe.iloc[1:, :]
        
        
        current_week_classification_dataframe = current_week_classification_dataframe.reset_index(drop =True)
        
        
        current_week_classification_dataframe.to_csv("files/" + c + "/"+ e + ".csv", index = False)
        
        #print(current_week_classification_dataframe)
        
        print(e, )
        
    
```
