```python
#this is the function that calculate the correlation of the data

def correlate(current_week):
    
        # declare the variable to hold all corelation data
    combine_correlation_data = pd.DataFrame()

    #loop through all currency
    for z in all_currency_pairs:

        current_currency = z
        raw_data = pd.read_csv('files/' + current_week + '/' + current_currency + '.csv'  ) 
        
        #print(z, raw_data)

        weeks = [50, 30, 20, 15, 10]

        #loop through all availale weeks
        for y in weeks:

            currency_list = []
            correlation_score = []

            #loops through all currency pairs again to calculate the corelation of each currency pair with the current iteration of z
            for x in all_currency_pairs:          

                #duplicates the raw data to work with
                current_data = raw_data.copy()

                #returns on the the amount of rows of weeks needed to perform correlation on
                current_data = current_data.head(y)

                #returns the two columns that correlation is being calculated for
                new_df = current_data[['nextweek_'+ current_currency, x]]
                
                #print(new_df)
                
                correlation = new_df.corr()
                
                #print(correlation)
                correlation = correlation[x][0:1].values[0]
                
                #print(correlation)

                #gets the absolute value of the correlation to 5 digit number
                correlation = abs(round(correlation, 5))
                
                
                #apppends the current currency and the corelation values to their respective lists
                currency_list.append(x)
                correlation_score.append(correlation)

            #creates a mini data frame with the information
            current_correlation_dataframe = pd.DataFrame()  
            current_correlation_dataframe [current_currency + '__' + str(y) ] = currency_list
            current_correlation_dataframe [current_currency + '_' + str(y) ] = correlation_score

            # sort the dataframe with corelation value from large to small
            current_correlation_dataframe.sort_values(by=[current_currency + '_' + str(y)], ascending=False, inplace=True, ignore_index= True)

            #fill the NaN values with zero
            current_correlation_dataframe.fillna(0, inplace=True)

            # add the current_currency dataframe to the global dataframe
            combine_correlation_data = pd.concat((combine_correlation_data, current_correlation_dataframe),axis=1)
            #print(len(correlation_score))

    #print(combine_correlation_data)
    
    #print(current_week)
    
    


    # save global dataframe to csv
    combine_correlation_data.to_csv('files/' + current_week + '/auto-correlation2.csv', index=False)



```

```python
# function that gets the needed features from the correlation information

def needed_features(processed_week):
    
    
    # load the corelation data from CSV
    correlation_dataframe = pd.read_csv('files/' + processed_week + '/auto-correlation2.csv'  ) 

    # declair the variable that will hold all testable units, the week and the needed feature
    currency_pair_detail = []
    
    # loop through all currency pairs
    for a in all_currency_pairs:

        current_currency = a.replace('10080','')


        weeks = [    {'week_name' : 10, 'correlation_cutoff' : 0.632},
                     {'week_name' : 15, 'correlation_cutoff' : 0.514},
                     {'week_name' : 20, 'correlation_cutoff' : 0.444},
                     {'week_name' : 30, 'correlation_cutoff' : 0.361},
                     {'week_name' : 50, 'correlation_cutoff' : 0.279},
                ]


        #loop throuh all available weeks
        for b in weeks:

            needed_column_list = []

            #check if the current currency in the current week has any currency above the correlation score of the week
            #this is to avoid running analysis for weeks with no correlation

            if  correlation_dataframe[current_currency + '_' + str(b['week_name'])][0] > b['correlation_cutoff']:

                #loop though weeks to get all the currency pairs in the weeks lesser than current week according to my strategy
                for c in weeks:

                    if c['week_name'] <= b['week_name']:

                        #gets the needed two needed columns
                        current_needed_columns = correlation_dataframe[[current_currency + '__' + str(c['week_name']), current_currency + '_' + str(c['week_name'])]] 

                        #get the rows that has corelation
                        current_needed_columns = current_needed_columns.loc[ (current_needed_columns[current_currency + '_' + str(c['week_name'])] >= c['correlation_cutoff'])]

                        #converts the currency pairs that passes to list
                        current_needed_columns = current_needed_columns[current_currency + '__' + str(c['week_name'])].values.tolist()

                        #loop through all the currency pairs that passes and append each currency pair to the list if it is not there already
                        for d in current_needed_columns :
                            if d not in needed_column_list:
                                needed_column_list.append(d) 


            #creates a dictionary with the needed info for the currency and week
            current_currency_detail = {'currency_name': current_currency, 'week': b['week_name'], 'needed_features':needed_column_list }

            #appends all non empty currency detail to the global list
            if len(current_currency_detail['needed_features']) > 0:
                currency_pair_detail.append(current_currency_detail)








    #print(currency_pair_detail)

    return currency_pair_detail


```

```python
def learn_best_trading_parameters(processed_week, currency_pair_detail ):
    
    for e in currency_pair_detail:
    
        #print (e['currency_name'])
        current_currency = e['currency_name']

      
        raw_data = pd.read_csv('files/' + processed_week + '/'+ e['currency_name'] + '.csv'   ) 
        weeks = e['week']

        number_of_accurate_prediction = 0
        number_of_trades = 0


        step_down = 5
        cut_off = 0
        n_estimators = 100
        min_samples_leaf = 1
        max_features = "auto"
        max_samples = None



        for v in range(step_down):




            data = raw_data.copy()
            data = data[v:v+weeks]
            #print(data['date'])


            y = data['nextweek_'+ current_currency]


            X = data.loc[:, e['needed_features'] ]






            X_train = X[1:]
            y_train = y[1:]
            X_test = X[:1]
            y_test = y[:1]




            clf=RandomForestRegressor(n_estimators=n_estimators, n_jobs =-1, min_samples_leaf=min_samples_leaf,
                                         max_features=max_features, max_samples = max_samples)

            clf.fit(X_train,y_train)
            y_pred=clf.predict(X_test)


            y_test = y_test.values




            if y_pred[0] > cut_off  or y_pred[0] < (-1*cut_off) :
                number_of_trades = number_of_trades + 1
                #print(y_pred[0], y_test[0])


                if (y_pred[0] > 0 and y_test[0] == 1) or (y_pred[0] < 0 and y_test[0]) == -1:
                    number_of_accurate_prediction = number_of_accurate_prediction +1



        e['accuracy_score'] =  100*number_of_accurate_prediction/number_of_trades
        #print(e['accuracy_score'])
        
    return currency_pair_detail



```

```python
def next_week_prediction(currency_pair_detail):
    
    next_week_trades =[]

    for f in all_currency_pairs:

        current_currency = f.replace('10080','')
        current_currency_list =[]
        current_currency_best_trade = None

        for g in currency_pair_detail:



            if g['currency_name'] == current_currency:

                current_currency_list.append(g)


        current_currency_list.sort(key=lambda x: x['week'], reverse=False)

        for h in current_currency_list:

            if h['accuracy_score'] > 99.9:

                current_currency_best_trade = h

                break


        if current_currency_best_trade == None:

            for i in current_currency_list:

                if i['accuracy_score'] > 79.9:

                    current_currency_best_trade = i

                    break


        if current_currency_best_trade != None:

            next_week_trades.append(current_currency_best_trade)




    #print(next_week_trades)  
    
    return next_week_trades


```

```python
def save_next_week_trades(processed_week, next_week_trades):
    
    # creating a dataframe for nextweek trades and saving it in csv

    name =[]
    week = []
    needed_features = []
    accuracy =[]

    for j in next_week_trades:

        name.append(j['currency_name'])
        week.append(j['week'])
        needed_features.append(j['needed_features'])
        accuracy.append(j['accuracy_score'])


    next_week_trades_dataframe = pd.DataFrame()
    next_week_trades_dataframe['names']=name
    next_week_trades_dataframe['week'] =week
    next_week_trades_dataframe['needed_features'] = needed_features
    next_week_trades_dataframe['accuracy']=accuracy


    print(next_week_trades_dataframe)

    next_week_trades_dataframe.to_csv('files/' + processed_week + '/next_week_trades.csv', index = False   ) 
    
    return 'OK'
```

```python

for g in available_weeks:

    currency_pair_detail = needed_features(g)

    currency_pair_detail = learn_best_trading_parameters(g, currency_pair_detail)

    next_week_trades = next_week_prediction(currency_pair_detail)

    final_output = save_next_week_trades(g, next_week_trades)

    print('------$$$$$$$$$$$$$$$$$-----------------------------------------------------------', g)
```

```python
def predict_nextweek_trades_direction(processed_week):
    next_week_prediction_score = []
    next_week_prediction =[]

    next_week_trades = pd.read_csv('files/' + processed_week + '/next_week_trades.csv'  ) 




    for index, row in next_week_trades.iterrows():



        current_currency = row.names
        raw_data = pd.read_csv('files/' + processed_week +'/' + current_currency + '.csv'    ) 
        weeks = row.week
        wanted_features = ast.literal_eval(row.needed_features)
        n_estimators = 100
        min_samples_leaf = 1
        max_features = "auto"
        max_samples = None




        raw_data = raw_data[:weeks]

        X_train = pd.DataFrame()

        for q in wanted_features:

            X_train[q] = raw_data[q]


        #print(X_train)  

        y_train = raw_data['nextweek_'+ current_currency]

        #print(y_train)



        prediction_model=RandomForestRegressor(
                                                    n_estimators=n_estimators,
                                                   n_jobs =-1,
                                                   min_samples_leaf=min_samples_leaf,
                                                    max_features=max_features,
                                                   max_samples = max_samples
                                              )



        prediction_model.fit(X_train,y_train)


        current_week_values = pd.read_csv('files/' + processed_week + '/current_week_values.csv'   ) 

        X_pred = pd.DataFrame()

        for h in wanted_features:

            X_pred[h] = current_week_values[h]

        #print(X_pred)

        y_pred = prediction_model.predict(X_pred)


        if y_pred[0] > 0:

            value = 'B'

        else:
            value = 'S'

        next_week_prediction_score.append(y_pred[0])
        next_week_prediction.append(value)

        print(row.names, row.week, y_pred)



    #print(next_week_prediction)
    #print(processed_week)
    
    next_week_trades['next_week_prediction_score'] = next_week_prediction_score
    next_week_trades['next_week_prediction'] = next_week_prediction

    next_week_trades.to_csv('files/' + processed_week + '/next_week_trades.csv', index = False )
```

```python
for g in available_weeks:
    predict_nextweek_trades_direction(g)
    print(g)
```
