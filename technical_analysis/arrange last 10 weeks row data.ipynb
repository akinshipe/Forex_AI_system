```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import metrics
import numpy as np

#it is important that the arrangement of the currencies when training the model in the training script
#is the same with the currency arrangement here also, to avoid passing wrong data 
currency_list = ['USDCHF10080', 'GBPUSD10080', 'EURUSD10080', 'USDJPY10080', 'USDCAD10080', 'AUDUSD10080', 'NZDUSD10080',
                 'GBPCHF10080', 'EURCHF10080', 'CHFJPY10080', 'CADCHF10080', 'AUDCHF10080', 'NZDCHF10080', 'EURGBP10080',
                 'GBPJPY10080', 'GBPCAD10080', 'GBPAUD10080', 'EURJPY10080', 'EURCAD10080', 'EURAUD10080', 'EURNZD10080',
                'CADJPY10080', 'AUDJPY10080', 'NZDJPY10080', 'AUDCAD10080', 'NZDCAD10080', 'AUDNZD10080']


```

```python
#get the last x timeframe from first currency(usdchf ) in currency list

#add names to columns in dataframe
headers = ['date_start','ignore', 'open', 'high', 'low', 'close', 'volume'  ]
recent_timeframes = pd.read_csv('recent_data/' + currency_list[0] + '.csv', names=headers  ) 
    
recent_timeframes = recent_timeframes['date_start'].values.tolist()# retrieve only the date_start column

first_ten_rows = slice(10)# used to create a slice object with 10 rows to use. check the python slice method

recent_timeframes = recent_timeframes[first_ten_rows]# slice object was passed to the list to grt the five members

recent_timeframes


```

```python
#loop through the whole currency and get the combine data for the needed row

combine_data_dataframe = pd.DataFrame(data={}, index = pd.Series(range(0,10)), )

# add the timeframe column to combined data
combine_data_dataframe['date_start'] = recent_timeframes


for x in currency_list :
    
    headers = ['date_start', 'ignore', x.replace('10080','') +'_open', x.replace('10080','') +'_high', x.replace('10080','')+'_low', x.replace('10080','')+'_close', x.replace('10080','') +'_volume'  ]
    current_currency_dataframe = pd.read_csv('recent_data/' + x + '.csv', names=headers  ) 
    
   
    
    #re-order columns in appropriate order and remove unneeded columns
    current_currency_dataframe = current_currency_dataframe[[x.replace('10080','') +'_volume',
                                                             x.replace('10080','') +'_volume',
                                                             x.replace('10080','') +'_volume',
                                                             x.replace('10080','') +'_volume',
                                                             x.replace('10080','') +'_volume',
                                                             x.replace('10080','') +'_volume',
                                                             
                                                             x.replace('10080','') +'_open',
                                                              x.replace('10080','') +'_open',
                                                              x.replace('10080','') +'_open',
                                                              x.replace('10080','') +'_open',
                                                              x.replace('10080','') +'_open',
                                                              x.replace('10080','') +'_open',
                                                             
                                                             x.replace('10080','') +'_high',
                                                              x.replace('10080','') +'_high',
                                                              x.replace('10080','') +'_high',
                                                              x.replace('10080','') +'_high',
                                                              x.replace('10080','') +'_high',
                                                              x.replace('10080','') +'_high',
                                                             
                                                             x.replace('10080','')+'_low', 
                                                               x.replace('10080','')+'_low', 
                                                               x.replace('10080','')+'_low', 
                                                               x.replace('10080','')+'_low', 
                                                               x.replace('10080','')+'_low', 
                                                               x.replace('10080','')+'_low', 
                                                             
                                                             x.replace('10080','')+'_close']]
    
    #print()
    #get only the top 10 columns
    current_currency_dataframe = current_currency_dataframe.head(10)   
    
    
    #add it to combine data_frame
    combine_data_dataframe = pd.concat([combine_data_dataframe, current_currency_dataframe], axis = 1 )  
    

combine_data_dataframe
```
