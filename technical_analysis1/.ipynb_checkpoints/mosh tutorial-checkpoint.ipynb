```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import metrics
import numpy as np




```

```python
data = pd.read_excel('currency_test.xlsx', sheet_name='USDCHF10080')
X = data.drop(columns=['date start', 'date end', 'next week close', 'next_week_close_pips'])
y = data['next_week_close_pips']
 
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, )
    
model = LinearRegression()
model.fit(X,y)
```

```python


# Create an empty list
Row_list =[]
  
# Iterate over each row
for index, rows in X.iterrows():
    # Create list for the current row
    my_list =[rows.volume, rows.open , rows.high, rows.low, rows.close]
      
    # append the list to the final list
    Row_list.append(my_list)
  

Row_list


```

```python
y_pred = model.predict(Row_list)
y_pred
```

```python
data['forecasted_pips'] = y_pred
print(data)
```

```python
print('MAE:', metrics.mean_absolute_error(y, y_pred))
print('MSE:', metrics.mean_squared_error(y, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y, y_pred)))
```
